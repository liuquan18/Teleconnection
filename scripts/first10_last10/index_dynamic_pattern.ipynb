```python
import xarray as xr
import numpy as np
```

```python
def post_process(data):
    data = data.rename({'__xarray_dataarray_variable__':'pc'})
    data['pc'].attrs = {'long_name':'principle component'}
    data['hlayers'].attrs = {'long_name':'geopotential height', "units":'pa'}
    data['mode'].attrs = {'long_name':'teleconnection modes'}
    data['ens'].attrs = {'long_name':'ensemble'}
    data['time'] = data.indexes['time'].to_datetimeindex()
    return data
```

# First10

```python
all_first = xr.open_dataset("/work/mh0033/m300883/3rdPanel/data/EOF_result/std_nonind_rolling_first/pc.nc")
```

```python
first_pc = all_first.isel(time = slice(0,10))
```

```python
first_pc = post_process(first_pc)
```

```python
first_pc.to_netcdf("/work/mh0033/m300883/3rdPanel/data/first_last/dep/first_pc.nc")
```
