# Read Data

The data to be averaged

```python
# %%
print("reading data...")
# data
allens = xr.open_dataset(
    "/work/mh0033/m300883/transition/gr19/gphSeason/allens_season_time.nc"
)
# split ens
splitens = tools.split_ens(allens)

# demean ens-mean
demean = splitens - splitens.mean(dim="ens")

# select traposphere
trop = demean.sel(hlayers=slice(20000, 100000))

trop = trop.var156
```

```python
first_NAO_lat_height = pattern_stat.lat_height(first_sptial_pattern,mode = 'NAO')
last_NAO_lat_height = pattern_stat.lat_height(last_sptial_pattern,mode = 'NAO')
```

```python
first_EA_lon_height = pattern_stat.lon_height(first_sptial_pattern,mode = 'EA')
last_EA_lon_height = pattern_stat.lon_height(last_sptial_pattern,mode = 'EA')
```

```python
fig, axes = pplt.subplots(nrows=2, ncols=2, wspace=2, hspace=2, sharex=False)
axes.format(
    abc="a",
    abcloc="ul",
    xminorlocator="null",
    yminorlocator="null",
    suptitle=f"spatial pattern change",
    yreverse=True,
    xreverse=False,
    toplabels=("meridional NAO", "horizontal EA"),
    leftlabels=("pos", "neg"),
)

# first10 years contourf
first_data_all =[first_NAO_lat_height.sel(extr_type = 'pos'),
                first_EA_lon_height.sel(extr_type = 'pos'),
                first_NAO_lat_height.sel(extr_type = 'neg'),
                first_EA_lon_height.sel(extr_type = 'neg')]

# last10 years
last_data_all =[last_NAO_lat_height.sel(extr_type = 'pos'),
                last_EA_lon_height.sel(extr_type = 'pos'),
                last_NAO_lat_height.sel(extr_type = 'neg'),
                last_EA_lon_height.sel(extr_type = 'neg')]

levels = [np.arange(-1.8, 1.9, 0.3),
          np.arange(-0.8, 0.9, 0.1),
          np.arange(-1.8, 1.9, 0.3),
          np.arange(-0.8, 0.9, 0.1),
        ]
shadings = []
lines = []
for ax, first_data, last_data,level in zip(axes,first_data_all, last_data_all,levels):
    # 
    shading = ax.contourf(
    first_data,
    x="lat",
    y="hlayers",
    levels=level,
    extend="both",
    )

    line = ax.contour(
    last_data,
    x="lat",
    y="hlayers",
    levels=level,
    extend="both",
    colors="k",
    color='gray8',
)
    ax.clabel(line,line.levels[::2])

    shadings.append(shading)
    lines.append(line)
    

axes[1, 0].format(xlabel="latittude")
axes[1, 1].format(xlabel="longitude")

axes[1, 0].colorbar(shadings[0], loc="b", label="NAO/std")
axes[1, 1].colorbar(shadings[1], loc="b", label="EA/std")

```
